```python
#monte carlo simulation to estimate pi
count1 = 0
count2 = 0
answer = 0
def randompoints(n):
    global count1
    global count2
    global answer
    for i in range(n):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        if(x**2+y**2<1):
            count1+=1
    return 4*(count1/n)
print(randompoints(10**7))


```

```python
string = "all happy families are alike each unhpapy family is unhappy in its own way"
def lettercount(string):
    dict = {}
    for i in string:
        if i in dict:
            dict[i] +=1
        else:
            dict[i] = 1
    return dict

print(lettercount(string))
```

```python
word = "It was the best of times, it was the worst of times... in short, the period was so far like the prsent period, that some of its noisiest authorities insisted on its being reieved, for good or for evil, in the superlative degree of comparison only."
def wordcount(word):
    count = 0
    for i in range(len(word)):
        if (word[i] == " "):
            count+=1
    if count>0:
        return count+1
    else:
        return 0

print(wordcount(word))
        
```

```python
def reverse_every_2nd(string):
    colors = string.split(" ")
    for i in range (len(colors)):
        if(i%2==1):
            colors[i] = colors[i][::-1]
        if(colors[i].find("i") != -1):
            colors[i] = colors[i][1:]
            colors[i] = colors[i] + ","
    display = " ".join(colors)
    return display

print(reverse_every_2nd(word))
```
