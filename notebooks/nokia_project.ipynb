```python
import os
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
import math

data_df = pd.read_csv("nok.csv")
# data_df.head()
all_df = data_df.drop(0)
# all_df.head()
# fig, ax = plt.subplots(figsize=(7, 4))

days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]
day_and_value = {}
for day in days:
    day_and_value[day] = all_df[all_df["Weekday"] == day]["Return"].mean()
print(day_and_value)
print(max(day_and_value, key = day_and_value.get))
print(min(day_and_value, key = day_and_value.get))

# Convert the Close column from strings to numbers, then multiply by 100
closing = pd.to_numeric(all_df["Close"], errors="coerce") * 100

# Make an array to count last digits 0-9
ends = np.zeros(10)

# Loop through each closing price
for i in range(len(closing)):

    # Skip missing/bad values
    if pd.notna(closing.iloc[i]):

        # Get the last digit after converting to cents
        last = int(round(closing.iloc[i])) % 10

        # Add 1 to that digit's count
        ends[last] += 1

# Divide by the number of valid closing prices to get proportions
final = ends / closing.notna().sum()

print(final)

```
