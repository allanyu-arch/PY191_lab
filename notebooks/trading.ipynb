```python
import pandas as pd
import numpy as np

# bollginer band --> if short-term > long-term --> buy
df1 = pd.read_csv("nok.csv")
my_money = 100
stock_money = 0
data = df1["Close", "Short_MA", "Long-MA"]

for i in range(len()):
    if data.iloc[i, 1] > data[i, 2]:
        stock_money -= data.iloc[i, 0]
    else:
        stock_money += data.iloc[i, 0]

print(my_money+stock_money)
```
